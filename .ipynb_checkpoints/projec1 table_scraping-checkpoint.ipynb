```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.iplt20.com/auction/2022"
r = requests.get(url)
print(r)
```

```python
soup = BeautifulSoup(r.text, "lxml")
print(soup)
```

```python
table = soup.find("table" , class_ = "ih-td-tab auction-tbl")
print(table)
```

```python
# find the title
title = table.find_all("th")
print(title)
```

```python
for i in title:
  print(i.text)
```

```python
# Another way to store the data into form of list
headers = []

for i in title:
  name = i.text
  headers.append(name)
headers
```

```python
# get the all rows

rows = table.find_all("tr")
print(rows)
```

```python
df = pd.DataFrame(columns = headers)
print(df)
```

```python
for i in rows[1:]:
  print(i)
```

```python
for i in rows[1:]:
  data = i.find_all("td")
  row = [tr.text for tr in data]
  print(row)
```

```python
for i in rows[1:]:
  first_td = i.find_all("td")[0].find("div", class_= "ih-pt-ic").text.strip()
  data = i.find_all("td")[1:]
  row = [tr.text for tr in data]
  row.insert(0, first_td)
  print(row)
```

```python
for i in rows[1:]:
  first_td = i.find_all("td")[0].find("div", class_= "ih-pt-ic").text.strip()
  data = i.find_all("td")[1:]
  row = [tr.text for tr in data]
  row.insert(0, first_td)
  # print(row)
  l = len(df)
  df.loc[l] = row
```

```python
print(df)
```

```python
df.to_csv("extracted data.csv")
```
