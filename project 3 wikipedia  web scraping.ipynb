```python
# import important libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/World_population"
r = requests.get(url)
print(r)
```

```python
# first way to get the html data 
soup = BeautifulSoup(r.text, "lxml")
print(soup)
```

```python
soup.title
```

```python
# Another way to fetch the html data 
soup1 = BeautifulSoup(r.content, "html.parser")
print(soup1)
```

```python
soup1.title
```

```python
soup1.title.text
```

```python
soup1.text
```

```python
# find all the tables

tables = soup.find_all("table")
print(tables)
```

```python
for i in tables:
    rows = i.find_all("tr")[1:]
    print(rows)
        
```

```python
for i in tables:
    rows = i.find_all("tr")[1:]

    data = []
    for row in rows:
        cols = row.find_all("td")
        cols = [col.text.strip() for col in cols]
        data.append(cols)
        print(data)
```

```python
for i in tables:
    rows = i.find_all("tr")[1:]

    data = []
    for row in rows:
        cols = row.find_all("td")
        cols = [col.text.strip() for col in cols]
        data.append(cols)
    df = pd.DataFrame(data)
    print(df)
    
```

```python
DataFrame = []
for i in tables:
    rows = i.find_all("tr")[1:]

    data = []
    for row in rows:
        cols = row.find_all("td")
        cols = [col.text.strip() for col in cols]
        data.append(cols)
    df = pd.DataFrame(data)
    DataFrame.append(df)
```
