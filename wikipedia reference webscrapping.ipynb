```python
# import important libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Check response
url = "https://en.wikipedia.org/wiki/Artificial_intelligence#References"
r = requests.get(url)
print(r)
```

```python
soup = BeautifulSoup(r.content, "html.parser")
soup
```

```python
ref = soup.find_all("div", class_ ="mw-references-wrap mw-references-columns")
ref
```

```python
span_element = soup.find_all("span", class_="reference-text")
span_element
```

```python
links = []
for i in span_element:
    a_elements = i.find_all("a")
    
    for a in a_elements:
        link =  a.get("href")
        print(link)
```
